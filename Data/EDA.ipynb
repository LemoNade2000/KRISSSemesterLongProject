```python
upordown = {}
for k in range(len(csv_list)): 
    read = pd.read_csv(csv_list[k])
    read = read.iloc[1:,]   #9월 21일에 열린 fomc 이후 한달치 데이터가 csv 파일에 없으므로, 해당 row 제외
    col1=read.iloc[:, 1:2]  
    filled=col1.notnull().sum()
    filled2=filled.loc['0']  #filled 2 는 해당 종목의 데이터가 있는 fomc의 수
    #filled2=filled.loc['0']
    upordown[csv_list[k][:-4]] = []
    for i in range(filled2):
        if read.iloc[i,1]>read.iloc[i,2]:
            upordown[csv_list[k][:-4]].append(0) #fomc전날에서 fomc 당일에 하락했다면 0, 상승했다면 1
        else:
            upordown[csv_list[k][:-4]].append(1)
print(upordown)
```
